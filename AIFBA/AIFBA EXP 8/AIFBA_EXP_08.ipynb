```python
# ===============================
# STEP 1: Import Libraries
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# For better visuals
sns.set(style="whitegrid")

# ===============================
# STEP 2: Load Dataset
# ===============================
# Use Kaggle dataset: creditcard.csv
# Upload in Colab before running

url = "https://storage.googleapis.com/download.tensorflow.org/data/creditcard.csv"
df = pd.read_csv(url)
df.head()

print("Dataset Shape:", df.shape)
print(df.head())

# ===============================
# STEP 3: Data Overview
# ===============================
print("\nClass Distribution:")
print(df['Class'].value_counts())

# ===============================
# STEP 4: Fraud vs Non-Fraud Count
# ===============================
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title("Fraud vs Non-Fraud Transactions")
plt.xticks([0,1], ['Non-Fraud', 'Fraud'])
plt.show()

# ===============================
# STEP 5: Transaction Amount Distribution
# ===============================
plt.figure(figsize=(8,5))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()

# ===============================
# STEP 6: Fraud Amount Distribution
# ===============================
plt.figure(figsize=(8,5))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Amount Distribution by Class")
plt.xticks([0,1], ['Non-Fraud', 'Fraud'])
plt.show()

# ===============================
# STEP 7: Correlation Heatmap
# ===============================
plt.figure(figsize=(12,10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()

# ===============================
# STEP 8: Time vs Fraud Pattern
# ===============================
plt.figure(figsize=(8,5))
sns.histplot(df[df['Class']==1]['Time'], bins=50, color='red')
plt.title("Fraud Transactions Over Time")
plt.show()

# ===============================
# STEP 9: Interactive Plot (Plotly)
# ===============================
fig = px.scatter(df.sample(2000),
                 x='Amount',
                 y='Time',
                 color='Class',
                 title="Fraud Detection Scatter Plot")
fig.show()

# ===============================
# STEP 10: Fraud Percentage
# ===============================
fraud_percentage = (df['Class'].sum() / len(df)) * 100
print(f"\nFraud Transactions Percentage: {fraud_percentage:.4f}%")

# ===============================
# STEP 11: Pairplot (Important Features)
# ===============================
sample_df = df[['V1', 'V2', 'V3', 'Amount', 'Class']].sample(1000)
sns.pairplot(sample_df, hue='Class')
plt.show()
```
